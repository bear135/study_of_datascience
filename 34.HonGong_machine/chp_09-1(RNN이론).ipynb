- Feedforward neural network : 입력데이터의 흐름이 앞으로만 전달됨(순서에 상관없음), CNN 
- Recurent neural network : 뉴런의 출력이 다시 자신에서 전달됨(바로 이전의 데이터를 재사용), RNN
- time step : 어떤 샘플을 처리하는 한단계. RNN은 이던 타임스탭을 기억하나, 스탭이 오래될 수록 기억이 희미해짐 
- tanh(hyperbolic tangent) : RNN에서 은닉층의 활성화 함수. 시그모이드와 비슷하나 값이 -1 ~ 1임  
- RNN의 출력층도 CNN과 마찬가지로 dense층을 사용. (2진분류: sigmoid, 다중분류: softmax) 
- 단 RNN의 출력층에서는 Flatten 클래스가 필요없다. CNN과 달리 마지막셀의 출력이 1차원이기 때문

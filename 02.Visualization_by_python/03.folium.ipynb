```python
m = folium.Map(location=[37.572, 127.001], zoom_start = 17)
tooltip = 'Click Please!'
folium.Marker([37.5720, 127.0008], popup = '<b>Doosan Digital Innovation</b>', 
              tooltip = tooltip).add_to(m)
m
```

```python
m = folium.Map(location=[37.572, 127.001], zoom_start = 17)
folium.Marker([37.5720, 127.0008], icon = folium.Icon(icon = 'cloud', color = 'red')).add_to(m)
m
```

```python
m = folium.Map(location=[37.572, 127.001], zoom_start = 17)
folium.Marker([37.5720, 127.0008], icon = folium.Icon(icon = 'info-sign', color = 'pink')).add_to(m)
m
```

```python
m = folium.Map(location=[37.572, 127.001], zoom_start = 17)
folium.Circle([37.57185, 127.00083], radius= 30, color='red', fill_color='red').add_to(m)
m
```

```python
m.add_child(folium.ClickForMarker(popup='Warning'))
m.add_child(folium.LatLngPopup())
m
```

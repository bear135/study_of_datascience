```python
#!pip install gdown
import pandas as pd 
import gdown  ## Google Drive file download PKG
gdown.download('https://bit.ly/3RhoNho', 'data/ns_202104.csv', quiet=False)
ns_df = pd.read_csv('data/ns_202104.csv', low_memory =False)

print(ns_df.shape)
ns_df.head()
```

```python
## 마지막 컬럼 제외 
ns_book = ns_df.iloc[:, :-1]
ns_book.head()
```

```python
## drop()을 사용하여 불필요한 컬럼 제거 
ns_book = ns_df.drop(['세트 ISBN', '부가기호', '권', '주제분류번호', 'Unnamed: 13'], axis = 1)
ns_book.head()
```

```python
## dropna()를 사용하여 NaN값이 포함된 컬럼 제거 
## how='all'을 지정하면 컬럼의 모든값이 NaN인 경우만 제거, 지정하지 않았을때는 NaN이 하나라도 존재하면 해당컬럼을 제거한다. 
ns_book = ns_df.dropna(axis=1)
ns_book.head()
```

```python
ns_book = ns_df.dropna(axis=1, how = 'all')
ns_book.head()
```

```python
## 특정 출판사의 책들만 추출 
ns_book2 = ns_df[ns_df['출판사'] == '한빛미디어']
print(ns_book2.shape)
ns_book2.head()
```

```python
## 대출 1000건 이상만 추출 
ns_book2 = ns_df[ns_df['대출건수'] >= 1000]
print(ns_book2.shape)
ns_book2.head()
```

```python
## duplicated()를 사용하여 중복 데이터 확인 
sum(ns_book.duplicated())
```

```python
## id기준으로는 중복이 없음, 도서명/저자/ISBN이 동일한 책들이 몇권이나 되는지 알아보자 
sum(ns_book.duplicated(subset = ['도서명', '저자', 'ISBN']))
```

```python
## - 중복이 있으면 처음과 마지막 값 중 무엇을 남길 것인가? : keep = 'first', 'last', False
## - 중복값 처리(unique한 1개의 key만 남기고 나머지 중복은 제거) : DataFrame.drop_duplicates()
ns_book3 = ns_book[ns_book.duplicated(subset = ['도서명', '저자', 'ISBN'], keep=False)]
ns_book3
```

---
- 인기도서 예측을 위해, 각 도서별로 대출건수를 합친다. 
1) 필요한 컬럼만 추출 
2) 중복처리를 위해 groupby() 사용
3) 주의 : 도서명, 저자 등에 NaN값 존재를 고려해야 함 (dropna = False)
---

```python
count_df = ns_book[['도서명', '저자', 'ISBN', '권', '대출건수']]
loan_cnt = count_df.groupby(by=['도서명', '저자', 'ISBN', '권'], dropna = False).sum()

print(loan_cnt.shape)
loan_cnt.head()
```

--- 
- 원본데이터에 대출건수 업데이트 하기 
1. 원본데이터의 중복처리 --> 고유한 행만 추출
2. 원본과 대출건수 데이터프레임의 인덱스 일치시키기 
3. update() 사용 
---

```python
## duplicated()로 중복된 행이 True 값을 가짐 --> ~로 반전시켜 고유한 행을 True로 변경 
dup_rows = ns_book.duplicated(subset=['도서명', '저자', 'ISBN', '권'])
unique_rows = ~dup_rows 
ns_book3 = ns_book[unique_rows].copy()

sum(ns_book3.duplicated(subset=['도서명', '저자', 'ISBN', '권']))
```

```python
## set_index()로 두 데이터의 인덱스를 일치시킨다. 
ns_book3.set_index(['도서명', '저자', 'ISBN', '권'], inplace = True)
ns_book3.head()
```

```python
## update()를 사용한다. 
ns_book3.update(loan_cnt)
ns_book3.head()
```

```python
## reset_index()로 인덱스 재설정 
ns_book4 = ns_book3.reset_index()
ns_book4.head()
```

```python
print('원본의 대출 100회이상 책 =', sum(ns_book['대출건수'] > 100))
print('중복처리후 대출 100회이상 책 =', sum(ns_book4['대출건수'] > 100))
```

- 중복처리를 통해 같은 책이 2번이상 나올 경우, 각각의 대출건수를 합한 값을 각 행에 할당해 주었기 때문에 숫자가 늘어났다. 

```python
## ns_book4를 원본데이터의 컬럼순서와 똑같게 수정한 후, csv로 저장 
ns_book4 = ns_book4[ns_book.columns]
ns_book4.head()
```

```python
ns_book4.to_csv('data/ns_book4.csv', index = False)
```

### 이상의 내용을 함수로 만들자 

```python
def data_cleaning(filename): 
    ns_df = pd.read_csv(filename, low_memory = False)
    ns_book = ns_df.dropna(axis = 1, how='all') ## 전체가 NaN인 컬럼은 제외 
    ## 중복을 고려하여, 책의 대출건수 구하기 
    count_df = ns_book[['도서명','저자','ISBN','권','대출건수']]
    loan_cnt = count_df.groupby(by=['도서명','저자','ISBN','권'], dropna=False).sum()
    ## 원본에서 중복제외 -> 고유한 행만 구하기 
    dup_rows = ns_book.duplicated(subset=['도서명','저자','ISBN','권'])
    unique_rows = ~dup_rows
    ns_book3 = ns_book[unique_rows].copy()
    ## 원본에 인덱스 설정 (업데이트할 대상의 인덱스와 일치)
    ns_book3.set_index(['도서명','저자','ISBN','권'], inplace = True)
    ns_book3.update(loan_cnt)
    ## 인덱스 재설정하고, 컬럼순서 원위치 
    ns_book4 = ns_book3.reset_index()
    ns_book4 = ns_book4[ns_book.columns]
    
    return ns_book4
```

```python
## 함수 실행 확인
new_ns_book = data_cleaning('data/ns_202104.csv')
print(new_ns_book.shape)
new_ns_book.head()
```

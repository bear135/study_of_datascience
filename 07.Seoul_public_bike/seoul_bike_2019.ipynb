```python
df2019 = pd.concat([df_201901, df_201902, df_201903, df_201904, df_201905, df_201906, df_201907, df_201908, df_201909])
df2019.info()
```

```python
df2019.isnull().sum()
```

```python
df2019['대여일시'] = pd.to_datetime(df2019['대여일시'])
df2019['반납일시'] = pd.to_datetime(df2019['반납일시'])
df2019.info()
```

```python
df2019['대여년'] = df2019['대여일시'].dt.year
df2019['대여월'] = df2019['대여일시'].dt.month
df2019['대여일'] = df2019['대여일시'].dt.day
df2019['대여시'] = df2019['대여일시'].dt.hour
df2019['대여분'] = df2019['대여일시'].dt.minute
df2019['대여요일'] = df2019['대여일시'].dt.dayofweek

df2019['반납년'] = df2019['반납일시'].dt.year
df2019['반납월'] = df2019['반납일시'].dt.month
df2019['반납일'] = df2019['반납일시'].dt.day
df2019['반납시'] = df2019['반납일시'].dt.hour
df2019['반납분'] = df2019['반납일시'].dt.minute
df2019['반납요일'] = df2019['반납일시'].dt.dayofweek

df2019.head(3)
```

```python
def concatenate_yy_mm_dd(대여일시) : 
    return '{0}-{1}-{2}'.format(대여일시.year, 대여일시.month, 대여일시.day)

df2019['대여연월일'] = df2019['대여일시'].apply(concatenate_yy_mm_dd)
df2019.head()
```

```python
def concatenate_yy_mm(대여일시) : 
    return '{0}-{1}'.format(대여일시.year, 대여일시.month)

df2019['대여연월'] = df2019['대여일시'].apply(concatenate_yy_mm)
df2019.head()
```

```python
## 월별, 요일별, 시간대별 자전거 대여량 추이 
df_mon = df2019['대여연월'].value_counts().reset_index()
df_mon.columns = ['대여연월', 'count']
print(df_mon.head())
plt.figure(figsize=(15, 5)) 
sns.lineplot(data=df_mon, x='대여연월', y='count')
```

```python
df_요일 = df2019['대여요일'].value_counts().reset_index()
df_요일.columns = ['대여요일', 'count']
print(df_요일.head())
plt.figure(figsize=(15, 5)) 
sns.barplot(data=df_요일, x='대여요일', y='count')
```

```python
df_hr = df2019['대여시'].value_counts().reset_index()
df_hr.columns = ['대여시', 'count']
print(df_hr.head())
plt.figure(figsize=(15, 5)) 
sns.barplot(data=df_hr, x='대여시', y='count')
```

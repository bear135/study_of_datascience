- file format transfer : json to csv

```python
import os
import json 
import glob 
import pandas as pd 
import pandas.io.json 
```

```python
project_list = []

## glob으로 result 폴더의 파일들을 읽는다. 와일드 카드를 사용할 수 있다. 
for filename in glob.glob('data2/result/*.json') : 
    project = json.loads(open(filename).read())
    project_list.append(project)
```

```python
## DataFrame 변환 (json_normalize 함수)
df = pandas.io.json.json_normalize(project_list) 

print(df.shape)
df.head(2)
```

```python
for col in df.columns: 
    print(col)
```

```python
## 컬럼명이 '_at'로 끝나는 것들은 unix_time 이다. 모두 datetime으로 바꿔준다. 

datetime_columns = filter(lambda a: a[-3:] == '_at', df.columns)

for column in datetime_columns : 
    df[column] = pd.to_datetime(df[column], unit = 's')
```

```python
print(df.shape)
df.head(1)
```

```python
## DataFrame to CSV 

csv_data = df.to_csv()
csv_data = csv_data.encode('cp949', 'ignore')    ### encoding 지정 

fp = open('data2/kickstarter_result.csv', 'wb')  ### 기존파일을 열고, csv 데이터를 쓰기한다. 
fp.write(csv_data)
fp.close()
```

```python
## 그룹핑하여 함수 적용하기 
tips = pd.read_csv('examples/tips.csv')
tips.head()
```

```python
tips['tip_ratio'] = tips['tip'] / (tips['total_bill']-tips['tip'])  ## tip의 비율 추가 
tips.head()
```

```python
tips_group = tips.groupby(['day', 'smoker'])
tips_group.sum()
```

```python
## group-by + aggregate 함수 --> 다수의 통계량을 한번에 ... 
tips_group.agg(['mean', 'median', 'min', 'max', 'std', 'sum'])
```

```python
tips_group['total_bill'].agg(['count', 'sum', 'mean', 'std'])
```

```python
## liner regression
import statsmodels.api as sm 
```

```python
def regress(data, yvar, xvar) : 
    y = data[yvar] 
    x = data[xvar]
    x['intercept'] = 1. 
    result = sm.OLS(y, x).fit()
    return result.params 
```

```python
## 인덱스 컬럼을 시간으로 인식시키기 
stock_data = pd.read_csv('examples/stock_px_2.csv', parse_dates = True, index_col = 0)
stock_data.info()
```

```python
stock_data.head()
```

```python
## S&P 500 지수(SPX)와 애플주식(AAPL)간의 선형회귀 
get_year = lambda x : x.year 
by_year = stock_data.groupby(get_year)

by_year.apply(regress, 'SPX', ['AAPL'])
```

```python
## Pandas Pivot_table
tips = pd.read_csv('examples/tips.csv')
tips['tip_ratio'] = tips['tip'] / (tips['total_bill']-tips['tip'])  ## tip의 비율 추가 
tips.head()
```

```python
tips.pivot_table(index = ['day', 'smoker'])
```

```python
## index, colums, contents를 지정하여 피봇테이블 만들기 
## margins = True ~ all(전체평균)을 생성 
tips.pivot_table(['size','tip_ratio' ], index=['time', 'day'], columns='smoker', margins=True)
```

```python
## aggfunc의 기본값은 mean, 다른 함수를 지정할 수 있다. 
tips.pivot_table(['size','tip_ratio' ], index=['time', 'day'], columns='smoker', margins=True, aggfunc='count')
```

```python
tips.pivot_table(['size','tip_ratio' ], index=['time', 'day'], columns='smoker', margins=True, aggfunc='sum')
```
